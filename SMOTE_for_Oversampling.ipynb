SMOTE (Chawla et al, 2022) can be used for oversampling as a method for combating imbalanced probelem. 
This code use Jason Brownlee website as reference.

https://dl.acm.org/doi/10.5555/1622407.1622416
https://machinelearningmastery.com/smote-oversampling-for-imbalanced-classification/

```python
# check version number
import imblearn
print(imblearn.__version__)
```

```python
# define dataset
import sklearn.datasets
X, y = sklearn.datasets.make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
```

```python
# summarize class distribution
from collections import Counter
counter = Counter(y)
print(counter)
```

```python
# scatter plot of examples by class label
from numpy import where
from matplotlib import pyplot
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

```python
# transform the dataset
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
# summarize the new class distribution
counter = Counter(y)
print(counter)
```

```python
# scatter plot of examples by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```
